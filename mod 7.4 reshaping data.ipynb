```python
import pandas as pd
long_df = pd.read_csv('data/long_data.csv',
usecols=['date', 'datatype', 'value']).rename(
columns={'value' : 'temp_C'}).assign(date=lambda x: pd.to_datetime(x.date),
temp_F=lambda x: (x.temp_C * 9/5) + 32)

long_df.head()
```

```python
long_df.head().T
```

```python
pivoted_df = long_df.pivot(index='date', columns='datatype', values='temp_C')

pivoted_df.head()
```

```python
pd.pivot(data=long_df,index="date",columns="datatype",values="temp_C").head()
```

```python
pivoted_df.describe()
```

```python
pivoted_df = long_df.pivot(index='date', columns='datatype', values=['temp_C', 'temp_F'])

pivoted_df.head()
```

```python
pivoted_df['temp_F']['TMIN'].head()
```

```python
multi_index_df = long_df.set_index(['date', 'datatype'])
multi_index_df.index
```

```python
multi_index_df.head()
```

```python
unstacked_df = multi_index_df.unstack()

unstacked_df.head()
```

```python
extra_data = pd.concat([long_df,
    pd.DataFrame([{'datatype': 'TAVG',
        'date': pd.Timestamp('2018-10-01'),
        'temp_C': 10,
        'temp_F': 50}])]).set_index(['date', 'datatype']).sort_index()

extra_data.head(8)

```

```python
extra_data.unstack().head()
```

```python
extra_data.unstack(fill_value=-40).head()
```

```python
wide_df = pd.read_csv('data/wide_data.csv')
wide_df.head()
```

```python
melted_df = wide_df.melt(id_vars='date',
value_vars=['TMAX', 'TMIN', 'TOBS'],
value_name='temp_C',
var_name='measurement')

melted_df.head()
```

```python
pd.melt(wide_df,
id_vars='date',
value_vars=['TMAX', 'TMIN', 'TOBS'],
value_name='temp_C',
var_name='measurement').head()
```

```python
wide_df.set_index('date', inplace=True)
wide_df.head()
```

```python
stacked_series = wide_df.stack()
stacked_series.head()
```

```python
stacked_df = stacked_series.to_frame('values')
stacked_df.head()
```

```python
stacked_df.index
```

```python
stacked_df.index.names
```

```python
stacked_df.index.rename(['date', 'datatype'], inplace=True)
stacked_df.index.names
```

```python
long_df.to_csv('data/long_df.csv')

pivoted_df.to_csv('data/pivoted_df.csv')

multi_index_df.to_csv('data/multi_index_df.csv')

unstacked_df.to_csv('data/unstacked_df.csv')

wide_df.to_csv('data/wide_df.csv')

melted_df.to_csv('data/melted_df.csv')

stacked_series.to_frame(name='value').to_csv('data/stacked_series.csv')
```
