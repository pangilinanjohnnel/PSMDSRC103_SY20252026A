```python
station_qm_wesf = df[df.station == '?'].WESF
```

```python
df.sort_values('station', ascending=False, inplace=True)
```

```python
df_deduped = df.drop_duplicates('date').drop(columns=['station', 'WESF']).sort_values('date').assign(WESF=station_qm_wesf)

df_deduped.shape
```

```python
df_deduped.dropna().shape
```

```python
df_deduped.dropna(how='all').shape
```

```python
df_deduped.dropna(how='all', subset=['inclement_weather', 'SNOW', 'SNWD']).shape
```

```python
df_deduped.dropna(axis='columns', thresh=df_deduped.shape[0]*.75).columns
```

```python
df_deduped.loc[:,'WESF'].fillna(0, inplace=True)
df_deduped.head()
```

```python
df_deduped.assign(
TMAX=lambda x: x.TMAX.replace(5505, np.nan).fillna(method='ffill'),
TMIN=lambda x: x.TMIN.replace(-40, np.nan).fillna(method='ffill')).head()
```

```python
df_deduped.assign(SNWD=lambda x: np.nan_to_num(x.SNWD)).head()
```

```python
df_deduped.assign(TMAX=lambda x: x.TMAX.replace(5505, np.nan).fillna(x.TMAX.median()),
TMIN=lambda x: x.TMIN.replace(-40, np.nan).fillna(x.TMIN.median()),
TOBS=lambda x: x.TOBS.fillna((x.TMAX + x.TMIN) / 2)).head()
```

```python
df_deduped.assign(TMAX=lambda x: x.TMAX.replace(5505, np.nan),
TMIN=lambda x: x.TMIN.replace(-40, np.nan)
).set_index('date').apply(lambda x: x.fillna(x.rolling(7, min_periods=0).median())).head(10)
```

```python
df_deduped.assign(TMAX=lambda x: x.TMAX.replace(5505, np.nan),
TMIN=lambda x: x.TMIN.replace(-40, np.nan),
date=lambda x: pd.to_datetime(x.date)
).set_index('date').reindex(
pd.date_range('2018-01-01', '2018-12-31', freq='D')
).apply(lambda x: x.interpolate()).head(10)
```

```python
df_deduped.to_csv('data/df_duped.csv')
```
